```python
import numpy as np
import pylab as plt
import matplotlib
import pandas as pd

from sklearn.metrics import mean_squared_error as mse
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import RidgeCV, LassoCV, ElasticNetCV
```

```python
df = pd.read_csv('./Steels_kaggle.csv')
df.head()
```

```python
X = df[[' C', ' Si', ' Mn', ' P', ' S', ' Ni', ' Cr', ' Mo',
       ' Cu', 'V', ' Al', ' N', ' Temperature (°C)']]
Y = df[' 0.2% Proof Stress (MPa)']
```

```python
pf = PolynomialFeatures(degree=3)
    
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
Xpoly_train = pf.fit_transform(X_train)
Xpoly_test = pf.fit_transform(X_test)
    
scaler = StandardScaler()
Xsc_train = scaler.fit_transform(Xpoly_train)
Xsc_test = scaler.fit_transform(Xpoly_test)
    
lassocv = LassoCV(alphas=np.linspace(1e-2,1e-1,100))
    
lassocv.fit(Xsc_train, y_train)
lassocv.alpha_
```

```python
fig = plt.figure(figsize=(10,7))

plt.plot(y_train, lassocv.predict(Xsc_train), 'or', label='Train')
plt.plot(y_test, lassocv.predict(Xsc_test), 'ob', label='Test')
plt.xlabel('Real value')
plt.ylabel('Predicted value')
plt.legend()
plt.show()
```

```python
mse_test = mse(y_test, lassocv.predict(Xsc_test), squared=False)
mse_train = mse(y_train, lassocv.predict(Xsc_train), squared=False)
print("Lasso: MSE for test is", mse_test, "and MSE for train is", mse_train)
```

```python
ridgecv = RidgeCV(alphas=np.linspace(0.1,1,100))
    
ridgecv.fit(Xsc_train, y_train)
ridgecv.alpha_
```

```python
fig = plt.figure(figsize=(10,7))

plt.plot(y_train, ridgecv.predict(Xsc_train), 'or', label='Train')
plt.plot(y_test, ridgecv.predict(Xsc_test), 'ob', label='Test')
plt.xlabel('Real value')
plt.ylabel('Predicted value')
plt.legend()
plt.show()
```

```python
mse_test = mse(y_test, ridgecv.predict(Xsc_test), squared=False)
mse_train = mse(y_train, ridgecv.predict(Xsc_train), squared=False)
print("Ridge: MSE for test is", mse_test, "and MSE for train is", mse_train)
```

```python
elasticcv = ElasticNetCV(alphas=np.linspace(1e-2,1e-1,100))
    
elasticcv.fit(Xsc_train, y_train)
elasticcv.alpha_
```

```python
fig = plt.figure(figsize=(10,7))

plt.plot(y_train, elasticcv.predict(Xsc_train), 'or', label='Train')
plt.plot(y_test, elasticcv.predict(Xsc_test), 'ob', label='Test')
plt.xlabel('Real value')
plt.ylabel('Predicted value')
plt.legend()
plt.show()
```

```python
mse_test = mse(y_test, elasticcv.predict(Xsc_test), squared=False)
mse_train = mse(y_train, elasticcv.predict(Xsc_train), squared=False)
print("ElasticNet: MSE for test is", mse_test, "and MSE for train is", mse_train)
```

### Implementation with Lasso is the best,  but takes a long time Let's take the Ridge

```python
for i in range(2, 8):
    pf = PolynomialFeatures(degree=i)
    
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
    Xpoly_train = pf.fit_transform(X_train)
    Xpoly_test = pf.fit_transform(X_test)

    scaler = StandardScaler()
    Xsc_train = scaler.fit_transform(Xpoly_train)
    Xsc_test = scaler.fit_transform(Xpoly_test)

    ridgecv = RidgeCV(alphas=np.linspace(0.1,10,100))

    ridgecv.fit(Xsc_train, y_train)
    ridgecv.alpha_
    mse_test = mse(y_test, ridgecv.predict(Xsc_test), squared=False)
    print("MSE for test with ridge for degree =", i, "is", mse_test)
```
